```python

# Load the pre-trained YOLOv8 model

model = YOLO('yolov8n.pt')

# Start training
model.train(data='data.yaml', imgsz=640, batch=16, epochs=100, device=0)

```

```python
import cv2
from matplotlib import pyplot as plt

# Load the trained model
model = YOLO(gdrive_path + "/runs/detect/train/weights/best.pt")

# Folder containing test images
image_folder = "test_images/"
output_folder = "predictions/"  # Folder to save output images
os.makedirs(output_folder, exist_ok=True)  # Create folder if it doesn't exist

# Get all image paths in the folder
image_paths = [os.path.join(image_folder, img) for img in os.listdir(image_folder) if img.endswith(('.png', '.jpg', '.jpeg'))]

# Process each image
for image_path in image_paths:
    # Predict on the image
    results = model.predict(source=image_path, imgsz=640, conf=0.5)

    # Visualize predictions
    results_image = results[0].plot()  # Render predictions on the image

    # Save the resulting image
    output_image_path = os.path.join(output_folder, os.path.basename(image_path))
    cv2.imwrite(output_image_path, results_image)  # Save to output folder

    # Display the image (optional, remove if processing many images to save time)
    plt.figure(figsize=(8, 8))
    plt.imshow(cv2.cvtColor(results_image, cv2.COLOR_BGR2RGB))
    plt.title(os.path.basename(image_path))
    plt.axis('off')
    plt.show()
```
